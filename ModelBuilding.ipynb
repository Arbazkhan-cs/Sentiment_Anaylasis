# Train Model

```python
from keras.models import Sequential
from keras.layers import Embedding, Dense, SimpleRNN

# Define the model
model = Sequential([
    Embedding(input_dim=50000, output_dim=2),
    SimpleRNN(32),
    Dense(4, activation='softmax')
])

# Compile the model
model.compile(loss='sparse_categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

# Print the model summary
model.summary()
```

```python
history = model.fit(X_train, y_train, epochs=10)
```

```python
history_ = model.fit(X_train, y_train, epochs=10)
```

```python
predict = model.predict(X_train)
```

# Save the Model And History

```python
model.save("/content/drive/MyDrive/Colab Notebooks/Sentiment Anyalasis/model.h5")
```

```python
loss1 = history.history['loss']
loss2 = history_.history['loss']

newLoss = loss1
for l in loss2:
  newLoss.append(l)

# Round each float value to the fourth decimal place
newLoss = [round(num, 4) for num in newLoss]
newLoss
```

```python
accuracy1 = history.history['accuracy']
accuracy2 = history_.history['accuracy']

newAccu = accuracy1
for ac in accuracy2:
  newAccu.append(ac)

# Round each float value to the fourth decimal place
newAccu = [round(num, 4) for num in newAccu]
newAccu
```

```python
newHistory = {'loss': [newLoss], 'accuracy': [newAccu]}
```
